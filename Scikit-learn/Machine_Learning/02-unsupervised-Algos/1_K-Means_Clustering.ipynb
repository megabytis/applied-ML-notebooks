- Clustering is a technique used in the K-means algorithm. In this algorithm, clustering refers to grouping similar data points based on their characteristics or features.

## K-Mean Algorithm:

- K mean is a clustering algorithm that is used to classify unlabeled data into groups/clusters based on similarity.

- Here K defines the number of pre-defined clusters that need to be created in the process if K=2, there will be two clusters, and for K=3, there will be three clusters, and so on.

- It is a centroid-based algorithm, where each cluster is associated with a centroid. The main aim of this algorithm is to minimize the sum of distances between the data point and their corresponding clusters.

- K-means then tries to determine different k-points called centroids, which are at the centre (least cumulative distance) from other points of the same class, but further away from points of another class.

## Why Do K-Means Clustering? (The Real Purpose):
- **To find hidden groups in your data when you don't have labels.**

## **Example:**
### Our Data Points:

Point A: (1, 1)

Point B: (2, 1)

## Why It's Called "K-Means"

- **K** = Number of clusters you want (2 in our example)
- **Means** = We keep moving centroids to the **mean** (average) of each cluster
- **K-Means** = Find K cluster centers by moving them to the means

---

## The Simple Algorithm:

1. **Pick K random points** as starting centroids
2. **For each data point**: Find closest centroid, assign to that cluster
3. **For each cluster**: Calculate mean of all points, move centroid there
4. **Repeat steps 2-3** until centroids stop moving

That's it! K-Means is literally just **"move to center, reassign, repeat"** until stable.

> **Remember**: K-Means is like herding cats ‚Äî you keep moving to the center of each group until everyone settles down! üê±

## Real-World Use Cases:

### 1. **Customer Segmentation**
- Group customers by purchasing behavior
- **Business value**: Target marketing campaigns to specific segments
- Example: "High-value customers", "Budget shoppers", "Occasional buyers"

### 2. **Image Compression**
- Group similar colors together
- **Technical value**: Reduce image file size by using fewer colors
- Example: 16M colors ‚Üí 64 representative colors

### 3. **Anomaly Detection**
- Find points that don't fit any cluster well
- **Security value**: Detect fraud, network intrusions, equipment failures
- Example: Credit card transactions that are far from normal spending clusters

### 4. **Market Research**
- Group similar products/reviews/features
- **Strategic value**: Identify product categories, customer preferences
- Example: Group smartphones by features/price/performance

### 5. **Document Organization**
- Group similar articles/news/social media posts
- **Content value**: Topic modeling, recommendation systems
- Example: News articles about "sports", "politics", "technology"

## Key Insight:
**You use K-Means when you want to discover structure in unlabeled data** ‚Äî not to predict something, but to understand what groups naturally exist.

> üéØ **Remember**: K-Means answers "What groups exist in my data?" not "What will happen next?" ‚Äî that's classification/regression.

# Let's Code :--

## STEP-1: Setting up 

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
````

````python
# Generating synthetic dataset
X, y = make_blobs(
    n_samples=1000,  # total points
    centers=4,  # like k=2 clusters
    cluster_std=0.8,  # how sread out clusters are
    random_state=42,
)
````

````python
# Plot the raw data
plt.scatter(X[:, 0], X[:, 1], s=40, color='gray')
plt.title("Raw Data Points (Before Clustering)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

> üëÜ here are natural clusters, but K-Means doesn't know that yet

## STEP-2: Applying K-Means

````python
model = KMeans(n_clusters=4, random_state=42)
model.fit(X)

# getting cluster assignments
y_pred = model.predict(X)

# getting centroid coordinats
centroids=model.cluster_centers_
print(f"CLuster Centers:\n {centroids}")
````

## STEP-3: Visualizing the clusters

````python
plt.figure(figsize=(6, 5))
plt.scatter(
    X[:, 0], X[:, 1], c=y_pred, cmap="viridis", s=40, alpha=0.7, label="Data Points"
)

plt.scatter(
    centroids[:, 0], centroids[:, 1], color="red", s=200, marker="X", label="Centroids"
)

plt.title("K-Means Clustering Result (K=4)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
````

## STEP-4: Evaluating with Silhouette Score

````python
from sklearn.metrics import silhouette_score

score = silhouette_score(X, y_pred)
print(f"Silhoutte score: {score:.3f}")
````

| Silhouette Score | Meaning            |
| ---------------- | ------------------ |
| 0.7‚Äì1.0          | üî• Strong clusters |
| 0.5‚Äì0.7          | üëç Good            |
| 0.25‚Äì0.5         | üòê Weak            |
| < 0.25           | üíÄ Poor            |


## OPTIONAL-Step (Trying different k-values)

````python
inertia_values=[]
k_values=range(1,7)

for k in k_values:
    km=KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertia_values.append(km.inertia_)

plt.plot(k_values,inertia_values, 'bx-')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (SSE)")
plt.title("Elbow Method for optimal K")
plt.show()
````
